```python
#remove outliers
def remove_outliers(df, q=0.05):
    upper = df.quantile(1-q)
    lower = df.quantile(q)
    mask = (df < upper) & (df > lower)
    return mask
```

```python
def rem_outliers_iqr(x):
    lower = x.quantile(0.25) -((x.IQR*1.5))
    upper = x.quantile(0.75)+((x.IQR)*1.5))
    fil = (x<upper) & (x>lower)
    return fil
```
