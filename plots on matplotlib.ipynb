```python
## compute the x and y coordinates for points on a sine curve

x = np.arange(0,4 * np.pi,0.1)
y = np.sin(x)

plt.plot(x, y, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=1)

plt.title("sine wave form")
```

```python
## compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,4 * np.pi,0.5)
y_sin = np.sin(x)
y_cos = np.cos(x)

## setup a subplot grid that has height 2 and width 1,
# and set the first such subplot as active

plt.subplot(2,1,1)

#make the first plot

plt.plot(x, y_sin, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=1)

plt.title("sine wave form")

#set the second subplot as active and make the second plot

plt.subplot(2,1,2)
plt.plot(x, y_cos, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=1)

plt.title("sine wave form")

#show the figure
plt.show()
```

```python
## bar plt

x= [2,8,10]
y= [11,16,9]

x2= [3,9,11]
y2= [6,15,7]

plt.bar(x,y, align = 'center')
plt.bar(x2,y2, color= 'g')

plt.title('Bar Graph')

plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
## Histogram ## major differnce among bar and histo is, histogram shows the density/count on y-axis

a = np.array([22,87,56,97,31,35,41,43,48,25,41,68,27,99,65])
plt.hist(a)
plt.title("histogram")
plt.show()
```

# Pie-chart

```python
# data to plt

labels = 'python','c++','Ruby','Java'
sizes  = [300,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
